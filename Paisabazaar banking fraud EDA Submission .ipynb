# **Project Summary -**


**Paisabazaar** is a financial services platform that assists customers in discovering, comparing, and applying for a wide range of banking and credit products. As an integral part of its services, Paisabazaar focuses on evaluating the creditworthiness of individuals, which plays a crucial role in both loan approval and effective risk management for financial institutions. **Creditworthiness** assessment helps lenders decide whether a customer is eligible for credit and under what terms, thereby reducing financial risk and improving decision accuracy.

A key metric used in this assessment process is the credit score, which represents an individual’s likelihood of repaying loans or maintaining credit balances responsibly. **Financial institutions rely heavily on credit scores to evaluate potential borrowers, as these scores summarize a person’s past financial behavior into a single, actionable value**. Accurate credit scoring enables lenders to distinguish between low-risk and high-risk customers, ensuring responsible lending while expanding access to credit for deserving individuals.

The classification of credit scores can significantly enhance Paisabazaar’s credit assessment processes. By categorizing individuals into different risk groups, **Paisabazaar can improve loan approval efficiency, reduce the probability of loan defaults, and offer more personalized financial** advice to customers. Such personalization not only benefits customers by matching them with suitable financial products but also strengthens trust and satisfaction with the platform. Moreover, better credit classification helps financial institutions optimize their lending strategies and maintain healthier loan portfolios.

In this context, analyzing and classifying credit scores based on customer data becomes essential. **Customer data includes a variety of financial and behavioral attributes such as income, credit card usage, existing credit balances, repayment history, and payment behavior**. These features collectively provide insights into an individual’s financial habits and repayment capacity. By leveraging data-driven approaches, Paisabazaar can develop predictive models that estimate credit scores more accurately and consistently.

The primary objective of this case study is to build a model that predicts the credit score of individuals based on these various features. Such a model can significantly enhance decision-making processes by providing data-backed recommendations. **Accurate predictions help financial institutions identify reliable borrowers, adjust interest rates appropriately, and minimize exposure to risky lending decisions**. At the same time, customers benefit from faster approvals, fairer credit evaluations, and access to financial products aligned with their credit profiles.

The dataset used in this study contains customer-level information, uniquely identified by record and customer IDs. These identifiers ensure that each customer’s financial profile is tracked accurately without duplication. The structured dataset supports systematic analysis and modeling, making it possible to identify meaningful patterns and relationships between customer attributes and credit scores.

Overall, this credit scoring initiative highlights the importance of data analytics in modern financial services. By transforming raw customer data into actionable insights, Paisabazaar can improve credit assessment accuracy, enhance customer experience, and support smarter financial decisions. The integration of predictive modeling into credit evaluation not only strengthens risk management practices but also contributes to a more inclusive and efficient financial ecosystem.

# **FRAUD ANALYSIS**

````python
# 1. STANDARDIZE DATA
# Convert credit score if categorical
credit_mapping = {'Poor': 1, 'Standard': 2, 'Good': 3}
df['Credit_Score'] = df['Credit_Score'].map(credit_mapping)

# Select features
risk_features = df[
    ['Credit_Score', 'Delay_from_due_date', 'Outstanding_Debt', 'Num_of_Delayed_Payment']
]

# Convert all to numeric safely
risk_features = risk_features.apply(pd.to_numeric, errors='coerce')
risk_features = risk_features.fillna(risk_features.mean())

# Scale
scaler = StandardScaler()
scaled_data = scaler.fit_transform(risk_features)

# 2. ISOLATION FOREST
iso = IsolationForest(
    n_estimators=100,
    contamination=0.05,   # assume 5% risky/fraud
    random_state=42
)

df['Fraud_Risk_Flag'] = iso.fit_predict(scaled_data)
df['Fraud_Risk_Flag'] = df['Fraud_Risk_Flag'].map({1: 0, -1: 1})

# 3. RISK SUMMARY
print("Fraud / High-Risk Distribution:")
print(df['Fraud_Risk_Flag'].value_counts())

# 4. VISUALIZATIONS

# Risk flag count
plt.figure()
df['Fraud_Risk_Flag'].value_counts().plot(kind='bar')
plt.title("High-Risk vs Normal Customers")
plt.xlabel("0 = Normal, 1 = High Risk")
plt.ylabel("Count")
plt.show()
````

**From the above chart it can be analyzed that the "High Risk" is very minimal as compared to "Normal" Customers.**
